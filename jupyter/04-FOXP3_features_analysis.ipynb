```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
path='./01-clincial_fea.csv'
df=pd.read_csv(path,index_col=0)
df=df[df['tumor_high'].notna()] 
# df
feature_index=[-8,-7,-4,-5,-2,-1]
features = df.iloc[:, feature_index]
z_scores = np.abs(stats.zscore(features))
threshold = 3 
mask = (z_scores < threshold).all(axis=1)
filtered_df = df[mask]
filtered_features = filtered_df.iloc[:, feature_index]

print("原始数据样本数：", df.shape[0])
print("剔除异常值后样本数：", filtered_df.shape[0])

scaler = StandardScaler()
scaled_features = scaler.fit_transform(filtered_features)

sse = []
k_range = range(1, 11)  
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_features)
    sse.append(kmeans.inertia_)
optimal_k=3
plt.figure(figsize=(8, 6))
plt.plot(k_range, sse, marker='o')
plt.axvline(x=optimal_k, color='red', linestyle='--', label=f'Optimal k = {optimal_k}')
plt.title('Elbow Method For Optimal k')
plt.xlabel('Number of clusters k')
plt.ylabel('SSE')
plt.xticks(k_range)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_curve, roc_auc_score
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize

data=filtered_df
Six_feature_lst = [
     'tumor_high', 'tumor_low', 
    'stroma_high','stroma_low','tumor_stroma_high', 'final_ratio'
]
fea=data.loc[:,Six_feature_lst].values
labels=np.array(data['Cluster'])
X_train, X_test, y_train, y_test = train_test_split(fea, labels, test_size=0.3, random_state=42)
model = LogisticRegression(solver='lbfgs', max_iter=1000, random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
report = classification_report(y_test, y_pred)
plt.figure(figsize=(6, 5))

sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=[r'$\mathit{FOXP3}$'+'_Sp'+str(i+1) for i in range(3)],
            yticklabels=[r'$\mathit{FOXP3}$'+'_Sp'+str(i+1) for i in range(3)])
plt.xlabel('Predicted Class')
plt.ylabel('True Class')
plt.title('Confusion Matrix')
plt.show()
print("Accuracy:", accuracy)
print("Confusion Matrix:\n", cm)
print("Classification Report:\n", report)

y_score = model.predict_proba(X_test)
y_test_bin = label_binarize(y_test, classes=[0, 1, 2])
n_classes = y_test_bin.shape[1]

fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

all_fpr = np.unique(np.concatenate([fpr[i] for i in range(n_classes)]))
mean_tpr = np.zeros_like(all_fpr)
for i in range(n_classes):
    mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])
mean_tpr /= n_classes

fpr["macro"] = all_fpr
tpr["macro"] = mean_tpr
roc_auc["macro"] = auc(fpr["macro"], tpr["macro"])

plt.figure(figsize=(8, 6))
plt.plot(fpr["macro"], tpr["macro"],
         label=f'Macro-average ROC (AUC = {roc_auc["macro"]:.2f})',
         color='navy', linestyle=':', linewidth=4)

colors = ['aqua', 'darkorange', 'cornflowerblue']
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=2,
             label='ROC for '+ r'$\mathit{FOXP3}$'+'_Sp'+str(i+1)+ f'(AUC = {roc_auc[i]:.2f})')

plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves for Three-class Logistic Regression')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```
